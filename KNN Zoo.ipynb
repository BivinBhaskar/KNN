```python
## we dont need the categorical column for classification ,remove column animal_name
```

```python
zoo1=zoo.drop(['animal name'],axis=1)
zoo1
```

```python
feat=zoo1.iloc[:,:-1]
feat
```

```python
tar=zoo1[['type']]
tar
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(feat,tar,train_size=0.8,random_state=50)
```

```python
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
# KNN
from sklearn.neighbors import KNeighborsClassifier
```

```python
# hyperparameter tuning
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(xtrain,ytrain)
```

```python
ypred=knn.predict(xtest)
```

```python
pd.DataFrame(ypred)
```

```python
knn.score(xtrain,ytrain)
```

```python
from sklearn.model_selection import GridSearchCV
n_neighbors=np.array(range(1,50))
params=dict(n_neighbors=n_neighbors)
n_neighbors
```

```python
gd=GridSearchCV(knn,params)
gd.fit(xtrain,ytrain)
```

```python
gd.best_params_
```

```python
# plotting accuracy
acc1=[]
acc2=[]

for i in range (1,20):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(xtrain,ytrain)
    acc1.append(knn.score(xtrain,ytrain))
    acc2.append(knn.score(xtest,ytest))
```

```python
plt.plot(range(1,20),acc1,color='blue')
plt.plot(range(1,20),acc2,color='red')
```
