```python
# feature scaling
from sklearn.preprocessing import StandardScaler
```

```python
st=StandardScaler()
```

```python
df=glass.iloc[:,:-1]
df
```

```python
df1=pd.DataFrame(st.fit_transform(df),columns=df.columns)
df1
```

```python
glass1=pd.concat([df1,glass['Type']],axis=1)
glass1
```

```python
glass1.head()
```

```python
## train test split
from sklearn.model_selection import train_test_split
```

```python
feat=glass1.iloc[:,:-1]
feat
```

```python
tar=glass1[['Type']]
tar
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(feat,tar,train_size=0.8,random_state=20)
```

```python
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
# KNN
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(xtrain,ytrain)
```

```python
ypred=knn.predict(xtest)
```

```python
pd.DataFrame(ypred)
```

```python
knn.score(xtrain,ytrain)
```

```python
## hyperparameter tuning
```

## grid search

```python
from sklearn.model_selection import GridSearchCV
n_neighbors=np.array(range(1,50))
params=dict(n_neighbors=n_neighbors)
n_neighbours
```

```python
gd=GridSearchCV(knn,params)
gd.fit(xtrain,ytrain)
```

```python
gd.best_params_
```

#### grid search method says that best value for hyperparameter is k=1

```python
# bulid a model with k=1
knn=KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(xtrain,ytrain)
```

```python
ypred1=knn.predict(xtest)
```

```python
pd.DataFrame(ypred1)
```

```python
knn.score(xtrain,ytrain)
```

```python
# plotting accuracy
acc1=[]
acc2=[]

for i in range (1,10):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(xtrain,ytrain)
    acc1.append(knn.score(xtrain,ytrain))
    acc2.append(knn.score(xtest,ytest))
```

```python
plt.plot(range(1,10),acc1,color='blue')
plt.plot(range(1,10),acc2,color='red')
```

```python
# from the plot we can infer that at k=1 both training and testing accuracy is high,so ideal value for k is 1
```
